```python
import json
def get_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
data = get_data("data.json")
data
```

```python
#Lets write code to analysis the given data
print("Users And Pages Information.")
def proper_data(data):
    print("User Data About There Connections And Liked Pages.\n")
    for user in data["users"]:
        print(f"The Users ID & Name is {user["id"]} : {user["name"]} And his/her friends are : {user["friends"]} And his/her Liked Pages are : {user["liked_pages"]}")
    print("Pages data.\n")
    for page in data["pages"]:
        print(f"The Page ID : {page['id']} And Name Of The Page is : {page['name']}")
proper_data(data)
        
```
