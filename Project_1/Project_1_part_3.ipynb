Part - 3 :- Implementation of People You May Know

```python
import json

def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)

def feature(user_id,data):
    userfriends = {}
    for user in data["users"]:
        userfriends[user["id"]] = set(user["friends"])

    if user_id not in userfriends:
        return []

    directfriends = userfriends[user_id]
    suggestions = {}

    for friend in directfriends:
        # For all friends of friend
        for mutual in userfriends[friend]:
            # if mutual id is not the user id and mutual is not there in direct friends
            if mutual != user_id and mutual not in directfriends:
                # count mutual friends 
                suggestions[mutual] = suggestions.get(mutual,0) + 1
    sorted_suggestions = sorted(suggestions.items(),key=lambda x: x[1],reverse=True)
    return [user_id for user_id,mutual_count in sorted_suggestions]
                


data = load_data("massive_data.json")
user_id = 10
recc = feature(user_id,data)
print(f"People You May Know {user_id} : {recc}")
```
