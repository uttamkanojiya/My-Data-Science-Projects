Part - 1 :- By the raw text data parse it to json so further operations can be taken  

```python
with open("finaldata.txt",encoding="utf-8") as f:
    data = f.read()
data
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if (len(c))>3]
chunks[3]
```

```python
def parse_data(chunk):
        chunk = chunk.strip()
        sep_chunks = chunk.split("\n")
        username = sep_chunks[0]
        no_of_posts = sep_chunks[1]
        no_of_followers = sep_chunks[2]
        no_of_following = sep_chunks[3]
        name = sep_chunks[4]
        if (len(sep_chunks) > 5):
            type_of_page = sep_chunks[5]
            bio = "\n".join(sep_chunks[6:])
        else:
            type_of_page = "Unknown"
            bio = ""
    #print(f"Username : {username}, No of Posts : {no_of_posts}, No of Followers : {no_of_followers}, No of Following : {no_of_following}, Name : {name}, Type Of Page : {type_of_page},Bio : {bio}. sep='\n'")
        return {"Username":username,"No of Posts":no_of_posts,"No of Followers":no_of_followers,"No of Following":no_of_following,"Name":name,"Type Of Page":{type_of_page},"Bio":bio}
#print(parse_data(chunks[10]))
    
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunks = parse_data(chunk)
    all_chunks.append(parsed_chunks)
print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks,indent=4,default=list)
print(s)
with open("proper_data.json","w",encoding="utf-8") as f:
    f.write(s)
```
