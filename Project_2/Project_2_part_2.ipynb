```python
import json
data = json.load(open("proper_data.json","r",encoding="utf-8"))
data
```

```python
def convert_post(data):
    #chunk = data
    for i in range(len(data)):
        #for No of Posts
        no_of_posts = str(data[i]["No of Posts"]).split(" posts")[0].replace(",","") #Done
        data[i]["No of Posts"] = int(no_of_posts)
    return data
    
def convert_followers(data):
    #for no of followers
    for i in range(len(data)):
        followers = str(data[i]["No of Followers"])
        no_of_followers = followers.split(" followers")[0].replace(",","").replace("K","")
        if ("K" in followers):
            no_of_followers = int(float(no_of_followers)*1000)
        data[i]["No of Followers"] = int(no_of_followers)
    return data

def convert_following(data):
    #for no of following
    for i in range(len(data)):
        no_of_following = str(data[i]["No of Following"]).split(" following")[0].replace(",","")
        data[i]["No of Following"] = int(no_of_following)
    return data
    
        
data1 = convert_post(data)
data2 = convert_followers(data)
data3 = convert_following(data)

#write the updated values back to the same file 
with open("proper_data.json","w",encoding="utf-8") as f:
    json.dump(data1,f,indent=4,ensure_ascii=False)

with open("proper_data.json","w",encoding="utf-8") as f:
    json.dump(data2,f,indent=4,ensure_ascii=False)

with open("proper_data.json","w",encoding="utf-8") as f:
    json.dump(data3,f,indent=4,ensure_ascii=False)
```
