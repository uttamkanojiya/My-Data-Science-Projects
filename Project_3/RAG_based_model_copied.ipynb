```python
import subprocess
import os
import whisper
import json
model = whisper.load_model('large-v2')
```

```python
files = os.listdir('videos')
for file in files:
  tutorial = file.split(' - ')[1]
  tutorial_splited = tutorial.split('.')[0]
  tutorial_num = tutorial_splited.split('#')[1]
  splited = file.split(' - ')[0]
  file_name = splited.split("  ")[0]
  #print(file_name,tutorial_num)
  lists = subprocess.run(['ffmpeg','-i',f'videos/{file}',f'audios/{tutorial_num}_{file_name}.mp3'])
  print(lists)
  
```

```python
input_file  = os.listdir('audios')
for audio in input_file:
    result = model.transcribe(audio =f'audios/{audio} ',language='hi',task='translate')
    print(result['segments'])
    chunks = []
    for segment in result['segments']:
        chunks.append({'number':tutorial_num,'title':file_name,'start':segment['start'],'end':segment['end'],'text':segment['text']})

    chunks_with_metadata = {'chunks':chunks,'text':result['text']}

    with open (f"jsons/{audio}.json","w") as f:
        json.dump(chunks_with_metadata,f)
```
