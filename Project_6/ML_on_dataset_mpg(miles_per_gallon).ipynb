```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
datasets = sns.get_dataset_names()
datasets
```

```python
miles_per_gallon = sns.load_dataset('mpg')
miles_per_gallon
```

```python
miles_per_gallon['horsepower'] = miles_per_gallon['horsepower'].fillna(miles_per_gallon['horsepower'].mean())
```

```python
miles_per_gallon.isnull().sum()
```

```python
miles_per_gallon.tail()
```

```python
copy_miles_per_gallon = miles_per_gallon.drop(['name','origin'],axis=1).copy()
copy_miles_per_gallon
```

```python
#using the StratifiedShuffleSplit to split the dataset properly
split = StratifiedShuffleSplit(n_splits=1,test_size=0.1,random_state=78)
for train_index,test_index in split.split(copy_miles_per_gallon,copy_miles_per_gallon['cylinders']):
  strat_train_set = copy_miles_per_gallon.loc[train_index]
  strat_test_set = copy_miles_per_gallon.loc[test_index]
```

```python
strat_train_set.shape
```

```python
labels = strat_train_set['mpg']
features = strat_train_set.drop(['mpg'],axis=1)
```

```python
labels
```

```python
#forming an pipeline so that the numerical value can be scale down to lower value for better processing
num_pipeline = Pipeline([
    ('scaler', StandardScaler())
])
prepared_data = num_pipeline.fit_transform(features)
```

```python
#Loading different models

#linear Model
linearmodel = LinearRegression()
linearmodel.fit(prepared_data,labels)

#Decision Tree Model
decisiontree= DecisionTreeRegressor()
decisiontree.fit(prepared_data,labels)

#random forest model
randomforest = RandomForestRegressor()
randomforest.fit(prepared_data,labels)
```

```python
#lets run the model on the test data

#linear Model
strat_test_set_features = strat_test_set.drop(['mpg'],axis=1)
prepared_test_data = num_pipeline.transform(strat_test_set_features)
predictions = linearmodel.predict(prepared_test_data)
print(predictions)
```

```python
#checking the accuracy of the linear model
accuracy = r2_score(strat_test_set['mpg'],predictions)
print(f"Accuracy (R2 Score) is: {accuracy:.2%}")
```

```python
#Decisiontree Model
strat_test_set_features = strat_test_set.drop(['mpg'],axis=1)
prepared_test_data = num_pipeline.transform(strat_test_set_features)
predictions_decisiontree = decisiontree.predict(prepared_test_data)
print(predictions_decisiontree)
```

```python
#checking the accuracy of the decisiontree model
accuracy = r2_score(strat_test_set['mpg'],predictions_decisiontree)
print(f"Accuracy (R2 Score) is: {accuracy:.2%}")
```

```python
#randomforest Model
strat_test_set_features = strat_test_set.drop(['mpg'],axis=1)
prepared_test_data = num_pipeline.transform(strat_test_set_features)
predictions_randomforest = randomforest.predict(prepared_test_data)
print(predictions_randomforest)
```

```python
#checking the accuracy of the randomforest model
accuracy = r2_score(strat_test_set['mpg'],predictions_randomforest)
print(f"Accuracy (R2 Score) is: {accuracy:.2%}")
```
